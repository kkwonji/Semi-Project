```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

data = []

for page in range(1, 201):
    url = f"https://finance.naver.com/item/news.naver?code=005930&page={page}"
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")

    iframe = soup.find("iframe", {"id": "news_frame"})
    iframe_src = iframe["src"]

    iframe_response = requests.get("https://finance.naver.com" + iframe_src)
    iframe_soup = BeautifulSoup(iframe_response.text, "html.parser")

    table = iframe_soup.find("table", class_="type5")
    rows = table.find_all("tr")

    for row in rows:
        title = row.find("td", class_="title")
        if title is not None:
            title_text = title.text.strip()
            info = row.find("td", class_="info")
            date = row.find("td", class_="date")
            info_text = info.text.strip() if info is not None else ""
            date_text = date.text.strip() if date is not None else ""
            data.append({"Page": page, "Title": title_text, "Info": info_text, "Date": date_text})

df = pd.DataFrame(data)
df.to_csv("output.csv", index=False)
```
